```python
NZero = 50
T = 100
k = math.log1p(2) / T
clusters = []
t = 0
while t<=1100:
    clusters.append(NZero * math.pow(e, k*t))
    t += 1
print(clusters)
```

```python
plt.plot(clusters)
plt.show(clusters)
```

```python
print(clusters[-1])
```
